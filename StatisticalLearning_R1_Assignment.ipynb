```python
import pandas as pd 
```

```python
#skipping the first row  by skiprows=1 and keeping the datatypes intact and removing the junk values by na_values = '-'
liga=pd.read_csv("Laliga.csv", skiprows=1,na_values = '-')
liga.head(10)
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
#===========================================================================================================
#replacing nan with 0 as '-' is already replaced as NAN while reading
liga.fillna(0,inplace=True)
liga.head(10)
```

```python
#Print all the teams which have started playing between 1930-1980. (5 points)
#==============================================================================
#keep first 4 character in debut column and converting the debut_year to numeric to compare as debut is of object type
Debut_year = pd.to_numeric(liga['Debut'].str[:4])
liga[(Debut_year >=1930) & (Debut_year <= 1980)][["Team","Debut"]]
```

```python
#Print the list of teams which came Top 5 in terms of points (2.5 points)
#====================================================================
new_liga=liga.loc[:]
#sorting in decreasing order 
new_liga.sort_values("Points", axis = 0, ascending = False).head(5)[["Team","Points"]] 
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
#Goal_diff_count = GoalsFor - GoalsAgainst
#==============================================================================
def Goal_diff_count(df):
    
    df["Goal_diff_count"]=df["GoalsFor"]-df["GoalsAgainst"]
    #maximum goal difference
    goal_difference= df.loc[df["Goal_diff_count"].idxmax()]
    print("Team with maximum goal difference is ",goal_difference["Team"],"and the diference is",goal_difference["Goal_diff_count"])
    #minimum goal difference
    goal_difference= df.loc[df["Goal_diff_count"].idxmin()]
    print("Team with minumum goal difference is ",goal_difference["Team"],"and the diference is",goal_difference["Goal_diff_count"])
    #returns all the teams with their Goal Differences
    return (df.loc[:,["Team","Goal_diff_count"]])

#calls the function Goal_diff_count
output_Goal_diff_count=Goal_diff_count(liga)
output_Goal_diff_count

```

```python
#Create a new column with name “Winning Percent” and append it to the data set (5 points)
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#find the top five teams with highest winning percentage
#========================================================

liga["Winning Percent"]=(liga["GamesWon"]/liga["GamesPlayed"])*100
#replacing nan with 0
liga["Winning Percent"].fillna(0,inplace=True)
print(liga.tail())
liga.nlargest(5,["Winning Percent"])[["Team","Winning Percent"]]
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
#=======================================================================================================
#grouping by BestPosition and sum the points. Used reset_index() to convert it back into a DataFrame
grouped = liga.groupby("BestPosition")["Points"].sum().reset_index()
grouped
```
